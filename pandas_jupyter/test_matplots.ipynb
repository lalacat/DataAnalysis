```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
y
```

```python
x = range(len(y))
plt.plot(x,y)
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight')
```

```python
plt.figure(figsize=(7,4))
plt.plot(y.cumsum(),'b',lw=2)
plt.plot(y.cumsum(),'rd')
plt.grid(True)
plt.axis('scaled')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test Plot')

```

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y,'b',lw=2)
plt.plot(y,'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test Plot')
```

```python
plt.figure(figsize=(7,4))
plt.plot(y[:,0],'b',lw=2,label='1st')
plt.plot(y[:,1],'b',lw=2,label='2st')
plt.plot(y,'ro')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test Plot')
```

```python
a = y[:,0]*100
b = y[:,1]
plt.figure(figsize=(7,4))
plt.plot(a,'b',lw=2,label='1st')
plt.plot(a,'ro')
plt.plot(b,'b',lw=2,label='2st')
plt.plot(b,'ro')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test Plot')
```

```python
fig,ax1 = plt.subplots() # 设置数据1为坐轴
plt.plot(a,'r',lw=2,label='1st')
plt.plot(a,'bd')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test Plot')

ax2 = ax1.twinx() # 设置数据2为右轴
plt.plot(b,'g',lw=2,label='2nd')
plt.plot(b,'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')

```

```python
plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(a,'r',lw=2,label='1st')
plt.plot(a,'bd')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Test Plot')

plt.subplot(212)
plt.plot(b,'g',lw=2,label='2nd')
plt.plot(b,'ro')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
y = np.random.standard_normal((1000,2))
```

```python
plt.figure(figsize=(7,5))
plt.plot(y[:,0],y[:,1],'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
c = np.random.randint(0,10,len(y))
plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],c=c,marker='o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
plt.figure(figsize=(7,5))
plt.hist(y,label=['1st','2nd'],bins=25)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
plt.figure(figsize=(7,5))
plt.hist(y,label=['1st','2nd'],color=['r','b'],stacked=True,bins=20)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
fig,ax = plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax,xticklabels=['1st','2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```
